# Data Profiling Report - Part 1.A: Dataset Structure

Dataset: French Road Safety Open Data, year 2024 (accidents corporels).

This notebook loads the four raw tables (`caract`, `lieux`, `vehicules`, `usagers`)
and documents their structure: column names, data types, and the semantic meaning
of each column based on the official data.gouv.fr documentation.

Missing values, consistency checks and the overall data quality summary are
covered separately in `02_data_quality.ipynb`.

```python
import pandas as pd
from pathlib import Path

BRONZE_DIR = Path("data/bronze")
FILES = ["caract-2024.csv", "lieux-2024.csv", "vehicules-2024.csv", "usagers-2024.csv"]
```

## Load the raw tables

```python
tables = {
    name: pd.read_csv(BRONZE_DIR / name, sep=";", encoding="utf-8", low_memory=False)
    for name in FILES
}

for name, df in tables.items():
    print(name, df.shape)
```

## Column inventory and data types

```python
def column_inventory(name, df):
    descriptions = COLUMN_DESCRIPTIONS.get(name, {})
    return pd.DataFrame({
        "column": df.columns,
        "dtype": [str(df[col].dtype) for col in df.columns],
        "description": [descriptions.get(col, "") for col in df.columns],
    })

inventories = {name: column_inventory(name, df) for name, df in tables.items()}
```

## Takeaways

Put together, these four tables form a relational structure built around
`Num_Acc`: `caract` describes the accident itself (when, where, under what
conditions), `lieux` describes the road it happened on, `vehicules` lists
every vehicle involved, and `usagers` lists every person involved along with
how badly they were hurt. Almost none of the columns are free text - they are
compact integer codes - which keeps the raw files small but means every
column is unreadable on its own without the description mapping built above.
That also explains why a code like `-1` ("not specified") can hide inside a
column without ever looking like a missing value: it is a perfectly valid
integer, just not a meaningful one. That is exactly the kind of thing the
data quality notebook checks for next.
